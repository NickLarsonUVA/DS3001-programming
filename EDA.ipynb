```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/data/airbnb_NYC.csv", encoding="latin1")
df.head()
```

```python
df['Price'].unique()
```

```python
x = df['Price'].str.replace(',', '')
x = pd.to_numeric(x, errors = 'coerce')
x.unique()
```

```python
sns.histplot(x)
```

```python
y = np.log(x)

sns.histplot(y)
```

```python
sns.kdeplot(y) #seaborn kde plot
```

```python
mean = print(y.mean())
print(y.var())
print(y.std())
pandas_kde_plot = y.plot.kde() #pandas kde plot
```
